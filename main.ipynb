```python
# Data import
import numpy as np
import pandas as pd

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

df = pd.read_excel("pcos_data_base_ml.xlsx")
```

```python
# Describe data 
df.describe()
```

```python
#Print data types
df.dtypes
```

```python
# % of missing values in each column
nan_ratio = (df.isna().mean()*100).sort_values(ascending=False)

missing_lte40 = nan_ratio[nan_ratio <= 40]
print(f"{len(missing_lte40)}/{len(df.columns)} columns have 40% or less missing values.")
missing_lte40
```

```python
# Bin missingness per column in 10% steps
bins = np.arange(0, 110, 10)
labels = [f"{b}-{b+10}%" for b in bins[:-1]]

missing_bins = pd.cut(
    nan_ratio,
    bins=bins,
    labels=labels,
    include_lowest=True,
    right=False
 )

missing_matrix = (
    pd.DataFrame({"missing_pct": nan_ratio, "bin": missing_bins})
    .sort_values("missing_pct", ascending=False)
 )

# Summary counts per bin
summary = missing_matrix["bin"].value_counts().sort_index()
display(summary)

# Columns grouped by missingness bin
bin_to_columns = (
    missing_matrix.groupby("bin").apply(lambda s: list(s.index)).to_frame("columns")
 )
display(bin_to_columns)
```

```python
# Distribution plots for columns with <= 80% missingness (2 plots wide)
import matplotlib.pyplot as plt

nan_ratio = df.isna().mean() * 100
cols_to_plot = nan_ratio[nan_ratio <= 80].index

if len(cols_to_plot) == 0:
    print("No columns found with <= 80% missingness.")
else:
    ncols = 2
    n = len(cols_to_plot)
    nrows = int(np.ceil(n / ncols))
    fig, axes = plt.subplots(nrows, ncols, figsize=(12 * ncols, 3.8 * nrows))
    axes = np.array(axes).reshape(-1)

    num_cols = set(df.select_dtypes(include=[np.number]).columns)

    for i, col in enumerate(cols_to_plot):
        ax = axes[i]
        series = df[col]
        if col in num_cols:
            ax.hist(series.dropna(), bins=30, edgecolor="black")
            ax.set_ylabel("Count")
        else:
            counts = series.astype("string").value_counts(dropna=False).head(10)
            y = np.arange(len(counts))
            ax.barh(y, counts.values)
            ax.set_yticks(y)
            ax.set_yticklabels(counts.index.astype(str))
            ax.invert_yaxis()
            ax.set_xlabel("Count")
        ax.set_title(f"{col} ({nan_ratio[col]:.1f}% missing)")

    for j in range(i + 1, len(axes)):
        axes[j].axis("off")

    fig.suptitle("Distributions for columns with <= 80% missingness", y=1.02)
    fig.tight_layout()
    plt.show()
```
