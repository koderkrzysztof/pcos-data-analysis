```python
# ============================================================================
# HELPER FUNCTIONS
# ============================================================================

def to_numeric_series(*args):
    """Converts multiple arguments to numeric pandas Series."""
    return [pd.to_numeric(arg, errors="coerce") for arg in args]


def safe_div(a, b):
    """Safe division avoiding division by zero errors."""
    a = pd.to_numeric(a, errors="coerce")
    b = pd.to_numeric(b, errors="coerce")
    with np.errstate(divide="ignore", invalid="ignore"):
        return a / b


def safe_log(x):
    """Safe logarithm for positive values only."""
    x = pd.to_numeric(x, errors="coerce")
    return np.log(x.where(x > 0))


def find_column(df: pd.DataFrame, keywords: list[str], exclude: list[str] = None):
    """Return first column name containing any keyword (case-insensitive), optionally excluding matches."""
    exclude = exclude or []
    for col in df.columns:
        low = col.lower()
        if any(ex in low for ex in exclude):
            continue
        if any(kw.lower() in low for kw in keywords):
            return col
    return None


def find_columns(df: pd.DataFrame, keywords: list[str]):
    """Return all column names containing any keyword (case-insensitive)."""
    return [col for col in df.columns if any(kw.lower() in col.lower() for kw in keywords)]


def ensure_glucose_mmol(glu: pd.Series) -> pd.Series:
    """Return glucose in mmol/L; if values look like mg/dL (median >50) convert dividing by 18."""
    g = pd.to_numeric(glu, errors="coerce")
    med = g.median(skipna=True)
    if pd.isna(med):
        return g
    if med > 50:  # probably mg/dL -> convert
        return g / 18.0
    return g


def ensure_glucose_mgdl(glu: pd.Series) -> pd.Series:
    """Return glucose in mg/dL; convert from mmol/L if median <= 30 by multiplying by 18."""
    g = pd.to_numeric(glu, errors="coerce")
    med = g.median(skipna=True)
    if pd.isna(med):
        return g
    if med <= 30:  # probably mmol/L -> convert
        return g * 18.0
    return g


# ============================================================================
# HORMONAL / HYPERANDROGENISM
# ============================================================================

def fai(total_testosterone_ng_ml: pd.Series, shbg_nmol_l: pd.Series) -> pd.Series:
    """
    Free Androgen Index (FAI) = (Total testosterone / SHBG) * 100
    Markers of hyperandrogenism.
    """
    
    t, s = to_numeric_series(total_testosterone_ng_ml, shbg_nmol_l)
    
    # Convert testosterone from ng/mL to nmol/L (1 ng/mL = 3.467 nmol/L) 
    testosterone_nmol_l = t * 3.467
    
    return safe_div(testosterone_nmol_l, s) * 100

def lh_fsh_ratio(lh: pd.Series, fsh: pd.Series) -> pd.Series:
    """
    LH/FSH ratio.
    Elevated ratio indicates ovulatory dysfunction.
    """
    return safe_div(*to_numeric_series(lh, fsh))


def amh_fsh_ratio(amh: pd.Series, fsh: pd.Series) -> pd.Series:
    """
    AMH/FSH ratio.
    Marker of PCOM-heavy reproductive subtype.
    """
    return safe_div(*to_numeric_series(amh, fsh))


def androstenedione_dheas_ratio(androstenedione: pd.Series, dheas: pd.Series) -> pd.Series:
    """
    Androstenedione / DHEAS ratio.
    Indicates ovarian vs adrenal androgen source.
    !IMPORTANT! Check units for both parameters.
    """
    return safe_div(*to_numeric_series(androstenedione, dheas))


def amh_elevated_flag(amh: pd.Series, threshold: float = 3.5) -> pd.Series:
    """
    Binary flag for AMH > threshold (ng/mL).
    AMH > 3.5 ng/mL suggests PCOM.
    """
    return pd.to_numeric(amh, errors="coerce") > threshold


# ============================================================================
# METABOLIC / GLYCEMIC
# ============================================================================

def homa_ir(glucose: pd.Series, insulin: pd.Series) -> pd.Series:
    """
    HOMA-IR = (Glucose [mmol/L] × Insulin [μU/mL]) / 22.5
    Insulin resistance marker.
    Automatically converts glucose from mg/dL to mmol/L if needed.
    """
    g = ensure_glucose_mmol(glucose)
    i = pd.to_numeric(insulin, errors="coerce")
    return (g * i) / 22.5


def tyg_index(triglycerides: pd.Series, glucose: pd.Series) -> pd.Series:
    """
    TyG = ln( (TG [mg/dL] * glucose [mg/dL]) / 2 )
    Dyslipidemia/insulin resistance marker.
    Automatically converts glucose to mg/dL if provided in mmol/L.
    """
    tg = pd.to_numeric(triglycerides, errors="coerce")
    glu_mgdl = ensure_glucose_mgdl(glucose)
    with np.errstate(divide="ignore", invalid="ignore"):
        return np.log((tg * glu_mgdl) / 2)


def quicki(insulin: pd.Series, glucose: pd.Series) -> pd.Series:
    """
    QUICKI = 1 / (log(Insulin [μU/mL]) + log(Glucose [mg/dL]))
    Insulin sensitivity inverse.
    Uses natural log; requires positive values.
    """
    ins = pd.to_numeric(insulin, errors="coerce")
    glu = ensure_glucose_mgdl(glucose)
    lg_ins = safe_log(ins)
    lg_glu = safe_log(glu)
    with np.errstate(divide="ignore", invalid="ignore"):
        return 1.0 / (lg_ins + lg_glu)


# ============================================================================
# LIPID
# ============================================================================

def tc_hdl_ratio(total_cholesterol: pd.Series, hdl: pd.Series) -> pd.Series:
    """
    TC/HDL ratio.
    Cardiovascular risk dyslipidemia marker.
    """
    return safe_div(*to_numeric_series(total_cholesterol, hdl))


def tg_hdl_ratio(triglycerides: pd.Series, hdl: pd.Series) -> pd.Series:
    """
    TG/HDL ratio.
    Small dense LDL marker.
    """
    return safe_div(*to_numeric_series(triglycerides, hdl))


def ldl_hdl_ratio(ldl: pd.Series, hdl: pd.Series) -> pd.Series:
    """
    LDL/HDL ratio.
    Atherosclerosis risk marker.
    """
    return safe_div(*to_numeric_series(ldl, hdl))


def non_hdl_cholesterol(total_cholesterol: pd.Series, hdl: pd.Series) -> pd.Series:
    """
    non-HDL-C = Total Cholesterol - HDL.
    Cardiovascular risk marker (all apoB particles).
    """
    tc, h = to_numeric_series(total_cholesterol, hdl)
    return tc - h


def atherogenic_index(triglycerides: pd.Series, hdl: pd.Series) -> pd.Series:
    """
    Atherogenic Index = ln(TG/HDL).
    Atherogenic dyslipidemia marker.
    """
    tg, h = to_numeric_series(triglycerides, hdl)
    with np.errstate(divide="ignore", invalid="ignore"):
        return np.log(tg / h)


# ============================================================================
# ADRENAL / OVARIAN
# ============================================================================

def circadian_cortisol_decline_percent(cortisol_8am: pd.Series, cortisol_11pm: pd.Series) -> pd.Series:
    """
    Circadian cortisol decline – percent drop from morning to evening.
    """
    am, pm = to_numeric_series(cortisol_8am, cortisol_11pm)
    return ((am - pm) / am) * 100

def circadian_cortisol_ratio(cortisol_11pm: pd.Series, cortisol_8am: pd.Series) -> pd.Series:
    """
    Circadian cortisol ratio – evening-to-morning cortisol ratio.
    """
    pm, am = to_numeric_series(cortisol_11pm, cortisol_8am)
    return pm / am

# ============================================================================
# INFLAMMATORY / HEMATOLOGY
# ============================================================================

def nlr(neutrophils: pd.Series, lymphocytes: pd.Series) -> pd.Series:
    """Neutrophil-to-Lymphocyte Ratio (NLR) - marker of chronic inflammation."""
    neu, lym = to_numeric_series(neutrophils, lymphocytes)
    return neu / lym

def plr(platelets: pd.Series, lymphocytes: pd.Series) -> pd.Series:
    """Platelet-to-Lymphocyte Ratio (PLR) - correlates with IR and hyperandrogenism."""
    plt, lym = to_numeric_series(platelets, lymphocytes)
    return plt / lym

def mpr(monocytes: pd.Series, platelets: pd.Series) -> pd.Series:
    """Monocyte-to-Platelet Ratio (MPR)."""
    mono, plt = to_numeric_series(monocytes, platelets)
    return mono / plt

# ============================================================================
# THYROID
# ============================================================================

def tsh_ft4_ratio(tsh_uIU_ml: pd.Series, ft4_pmol_L: pd.Series) -> pd.Series:
    """TSH to FT4 ratio - marker of subclinical hypothyroidism."""
    tsh, ft4 = to_numeric_series(tsh_uIU_ml, ft4_pmol_L)
    return tsh / (ft4 * 12.87)  # Convert FT4 from ng/dL to pmol/L for ratio

def anti_thyroid_score(anty_tpo: pd.Series, anty_tg: pd.Series) -> pd.Series:
    """Composite score for thyroid autoimmunity."""
    atpo, atg = to_numeric_series(anty_tpo, anty_tg)
    return atpo + atg

# ============================================================================
# OTHER COMPOSITES
# ============================================================================

def ast_alt_ratio(ast: pd.Series, alt: pd.Series) -> pd.Series:
    """AST to ALT ratio - marker of NAFLD in insulin resistance."""
    a, b = to_numeric_series(ast, alt)
    return a / b

# ============================================================================
# MASTER FUNCTION: COMPUTE ALL INDICATORS
# ============================================================================
def compute_all_indicators(df: pd.DataFrame, glucose_unit: str = None) -> pd.DataFrame:
    """
    Computes all available indicators for the given DataFrame.
    Returns a copy of the DataFrame with new indicator columns added.
    """
    df = df.copy()
    
    # ===== HORMONAL / HYPERANDROGENISM =====
    if 'testosterone_ng_ml' in df.columns and 'shbg_nmol_l' in df.columns:
        df["FAI"] = fai(df['testosterone_ng_ml'], df['shbg_nmol_l'])
    
    if 'lh_iu_l' in df.columns and 'fsh_iu_l' in df.columns:
        df["LH_FSH_ratio"] = lh_fsh_ratio(df['lh_iu_l'], df['fsh_iu_l'])
    
    if 'AMH_ng_ml' in df.columns and 'fsh_iu_l' in df.columns:
        df["AMH_FSH_ratio"] = amh_fsh_ratio(df['AMH_ng_ml'], df['fsh_iu_l'])
    
    if 'androstenedione_ng_ml' in df.columns and 'dheas_ug_dL' in df.columns:
        df["Andro_DHEAS_ratio"] = androstenedione_dheas_ratio(df['androstenedione_ng_ml'], df['dheas_ug_dL'])
    
    if 'AMH_ng_ml' in df.columns:
        df["AMH_elevated"] = amh_elevated_flag(df['AMH_ng_ml'])
    
    # ===== METABOLIC / GLYCEMIC =====
    
    if 'glucose_fasting_mg_dL' in df.columns and 'insulin_fasting_uU_ml' in df.columns:
        df["HOMA_IR"] = homa_ir(df['glucose_fasting_mg_dL'], df['insulin_fasting_uU_ml'])
        df["QUICKI"] = quicki(df['insulin_fasting_uU_ml'], df['glucose_fasting_mg_dL']) 

    # TyG index
    if "Triglicerydy (TG)" in df.columns and 'glucose_fasting_mg_dL' in df.columns:
        df["TyG_index"] = tyg_index(df['Triglicerydy (TG)'], df['glucose_fasting_mg_dL'])
    
    # ===== LIPID =====
    if "Cholesterol całkowity (TCHOL)" in df.columns and "HDL cholesterol (HDL)" in df.columns:
        df["TC_HDL_ratio"] = tc_hdl_ratio(df['Cholesterol całkowity (TCHOL)'], df['HDL cholesterol (HDL)'])
        df["Non_HDL_C"] = non_hdl_cholesterol(df['Cholesterol całkowity (TCHOL)'], df['HDL cholesterol (HDL)'])

    if "Triglicerydy (TG)" in df.columns and "HDL cholesterol (HDL)" in df.columns:
        df["TG_HDL_ratio"] = tg_hdl_ratio(df['Triglicerydy (TG)'], df['HDL cholesterol (HDL)'])
        df["Atherogenic_Index"] = atherogenic_index(df['Triglicerydy (TG)'], df['HDL cholesterol (HDL)'])

    if "LDL cholesterol (LDL)" in df.columns and "HDL cholesterol (HDL)" in df.columns:
        df["LDL_HDL_ratio"] = ldl_hdl_ratio(df['LDL cholesterol (LDL)'], df['HDL cholesterol (HDL)'])
    
    # ===== INFLAMMATORY / HEMATOLOGY =====
    if "neutrophils_abs_10_3_uL" in df.columns and "lymphocytes_abs_10_3_uL" in df.columns:
        df['NLR'] = nlr(df['neutrophils_abs_10_3_uL'], df['lymphocytes_abs_10_3_uL'])
    
    if "platelets_10_3_uL" in df.columns and "lymphocytes_abs_10_3_uL" in df.columns:
        df['PLR'] = plr(df['platelets_10_3_uL'], df['lymphocytes_abs_10_3_uL'])
    
    # Monocyte-to-Platelet Ratio
    if "Morfologia krwi, z pełnym różnicowaniem granulocytów (leukocytów) z rozmazem - 5 Diff (MON#)" in df.columns and "platelets_10_3_uL" in df.columns:
        df['MPR'] = mpr(df['Morfologia krwi, z pełnym różnicowaniem granulocytów (leukocytów) z rozmazem - 5 Diff (MON#)'], df['platelets_10_3_uL'])
        
    # Circadian cortisol (using KOR and KOR23)
    if "cortisol_8am_ug_dL" in df.columns and "Kortyzol godz. 23:00 (KOR23)" in df.columns:
        df['Circadian_Cortisol_Decline_pct'] = circadian_cortisol_decline_percent(df['cortisol_8am_ug_dL'], df['Kortyzol godz. 23:00 (KOR23)'])
        df['Circadian_Cortisol_Ratio'] = circadian_cortisol_ratio(df['Kortyzol godz. 23:00 (KOR23)'], df['cortisol_8am_ug_dL'])
        
    # ===== THYROID =====
    if "tsh_uIU_mL" in df.columns and "FT4 (FT4)" in df.columns:
        df['TSH_FT4_ratio'] = tsh_ft4_ratio(df['tsh_uIU_mL'], df['FT4 (FT4)'])
    
    # Anti-thyroid score (combining multiple antibody columns)
    if 'Anty-TPO (ATA_TPO)' in df.columns and 'anti_tg_IU_mL' in df.columns:
        df['Anti_Thyroid_Score'] = anti_thyroid_score(df['Anty-TPO (ATA_TPO)'], df['anti_tg_IU_mL'])
    
    # ===== LIVER / NAFLD =====
    if 'ASPAT (AST)' in df.columns and 'ALAT (ALT)' in df.columns:
        df['AST_ALT_ratio'] = ast_alt_ratio(df['ASPAT (AST)'], df['ALAT (ALT)'])
    
    return df
```

```python
# Compute all indicators
df_with_indicators = compute_all_indicators(df)

# Get only the newly added columns (exclude original columns)
original_columns = set(df.columns)
new_columns = [col for col in df_with_indicators.columns if col not in original_columns]

if 'group' in df.columns:
    new_columns_with_group = ['group'] + new_columns
else:
    new_columns_with_group = new_columns
    print("Warning: No 'group' column found in original DataFrame")

# Display head of new columns
print("=" * 80)
print("HEAD OF NEWLY ADDED INDICATORS")
print("=" * 80)
display(df_with_indicators[new_columns_with_group].head(100))

# Display mean values by group
if 'group' in df.columns:
    print("\n" + "=" * 80)
    print("MEAN VALUES BY GROUP")
    print("=" * 80)
    
    # Mean for each group for all new columns
    means_by_group = df_with_indicators.groupby('group')[new_columns].mean()
    display(means_by_group)
    
    # Optional: Also display standard deviation
    print("\n" + "=" * 80)
    print("MEAN ± STANDARD DEVIATION BY GROUP")
    print("=" * 80)
    
    for group in df_with_indicators['group'].unique():
        print(f"\n--- {group} ---")
        for col in new_columns:
            mean_val = df_with_indicators[df_with_indicators['group'] == group][col].mean()
            std_val = df_with_indicators[df_with_indicators['group'] == group][col].std()
            print(f"{col}: {mean_val:.2f} ± {std_val:.2f}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# ============================================================================
# CORRELATION WITH 'group' (PCOS vs Control)
# ============================================================================

print("=" * 80)
print("CORRELATION WITH 'group' (PCOS vs Control_no_PCOS)")
print("=" * 80)

# Ensure 'group' is numeric for correlation (PCOS = 1, Control_no_PCOS = 0)
df_corr = df_with_indicators.copy()
df_corr['group_numeric'] = (df_corr['group'] == 'PCOS').astype(int)

# Get all numeric columns (excluding original group column)
numeric_cols = df_corr.select_dtypes(include=['float64', 'int64']).columns.tolist()
numeric_cols = [col for col in numeric_cols if col not in ['group_numeric']]

# Calculate correlation with group
group_correlations = []
for col in numeric_cols:
    corr_val = df_corr[col].corr(df_corr['group_numeric'])
    if not pd.isna(corr_val):
        group_correlations.append({
            'Parameter': col,
            'Correlation_with_PCOS': corr_val,
            'Absolute_Correlation': abs(corr_val)
        })

# Create DataFrame and sort by absolute correlation
group_corr_df = pd.DataFrame(group_correlations).sort_values('Absolute_Correlation', ascending=False)

print(f"\nAnalyzed {len(group_corr_df)} parameters for correlation with PCOS status")
print(f"\nTop 20 parameters most strongly associated with PCOS:")

# Display top 20
display(group_corr_df.head(20))

# ============================================================================
# VISUALIZATION: Bar plot of top 15 correlations
# ============================================================================

plt.figure(figsize=(12, 8))

top_corr = group_corr_df.head(15).copy()
top_corr = top_corr.sort_values('Correlation_with_PCOS', ascending=True)

colors = ['steelblue' if x < 0 else 'salmon' for x in top_corr['Correlation_with_PCOS']]

plt.barh(top_corr['Parameter'], top_corr['Correlation_with_PCOS'], color=colors, edgecolor='black')
plt.axvline(x=0, color='black', linewidth=0.8)
plt.xlabel('Correlation with PCOS (positive = higher in PCOS, negative = higher in Control)', fontsize=12)
plt.title('Parameters Most Strongly Associated with PCOS Status', fontsize=14)
plt.tight_layout()
plt.show()

# ============================================================================
# HEATMAP: Top correlations with group
# ============================================================================

plt.figure(figsize=(10, 6))

# Create a small DataFrame with top parameters
top_15_params = group_corr_df.head(15)['Parameter'].tolist()
corr_subset = df_corr[top_15_params + ['group_numeric']].corr()

# Extract only correlations with group_numeric
group_corr_subset = corr_subset[['group_numeric']].drop('group_numeric').sort_values('group_numeric')

# Plot heatmap
sns.heatmap(
    group_corr_subset,
    annot=True,
    fmt='.3f',
    cmap='RdBu_r',
    center=0,
    cbar_kws={"shrink": 0.8},
    square=True
)
plt.title('Correlation of Top 15 Parameters with PCOS Status', fontsize=14)
plt.ylabel('Parameter', fontsize=12)
plt.xlabel('Correlation with PCOS (1=PCOS, 0=Control)', fontsize=12)
plt.tight_layout()
plt.show()

# ============================================================================
# SUMMARY: Which parameters are higher in PCOS vs Control?
# ============================================================================
print("\n" + "=" * 80)
print("SUMMARY: Parameters that differ between PCOS and Control")
print("=" * 80)

higher_in_pcos = group_corr_df[group_corr_df['Correlation_with_PCOS'] > 0.1].sort_values('Correlation_with_PCOS', ascending=False)
higher_in_control = group_corr_df[group_corr_df['Correlation_with_PCOS'] < -0.1].sort_values('Correlation_with_PCOS', ascending=True)

print(f"\n📈 Parameters HIGHER in PCOS (|r| > 0.1): {len(higher_in_pcos)}")
for _, row in higher_in_pcos.head(10).iterrows():
    print(f"    {row['Parameter']}: {row['Correlation_with_PCOS']:.3f}")

print(f"\n📉 Parameters HIGHER in Control (|r| > 0.1): {len(higher_in_control)}")
for _, row in higher_in_control.head(10).iterrows():
    print(f"    {row['Parameter']}: {row['Correlation_with_PCOS']:.3f}")

# Save to CSV
group_corr_df.to_csv('correlation_with_group.csv', index=False)
print("\n✅ Saved to 'correlation_with_group.csv'")
```
